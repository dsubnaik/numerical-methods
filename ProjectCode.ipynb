Imports

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#read in file
df = pd.read_csv('cc1.csv')
```

see all column headers

```python
df.columns
```

split data frame

```python
y = df['Weight']
x = df[['cc_0412', 'cc_0415', 'cc_0427', 'cc_0506', 'cc_0516',
       'cc_0520', 'cc_0523', 'cc_0527', 'cc_0531', 'cc_0602', 'cc_0607',
       'cc_0614', 'cc_0617', 'cc_0620', 'cc_0623', 'cc_0627', 'cc_0630',
       'cc_0708', 'cc_0713', 'cc_0716', 'cc_0719', 'cc_0721', 'cc_0725',
       'cc_0728', 'cc_0802', 'cc_0808', 'cc_0812']]
```

split data into training and testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

```python
y_train = np.expand_dims(y_train, -1)
#print(y_train.shape)

y_test = np.expand_dims(y_test, -1)
#print(y_test.shape)
```

modeling and training

```python
import numpy as np
from numpy.linalg import inv

x = x_train

y = y_train

a = np.dot(x.T, x)
b = np.dot(x.T, y)
# print(A)

a_inv = inv(a)

# print(np.dot(A_inv, A))
sol = np.dot(a_inv, b)

print(sol)
```

```python
y_pred = np.dot(x_test, sol)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

Evaluation

```python
mybestmodel = LinearRegression()
mybestmodel.fit(x_train, y_train)

mybestmodel.score(x_test, y_test)


```
